```python
import matplotlib.pyplot as plt

x=[2,4,6,8,10]
y=[6,7,8,2,4]

plt.bar(x,y, label="Bars1")
plt.xlabel('x')
plt,ylabel('y')
plt.title('Interesting graph\nChek it out')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[4,5,6,7]
playing=[8,5,7,8,13]
slices=[7,2,2,13]
activities=['Massachusetts','University of Colifornia', 'Stanford', 'Harvard University' ]
cols=['c','m', 'r', 'b']
plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90,
       shadow= True,
       explode=(0,0.1,0,0,),
       autopct='%1.1f%%')
plt.title('World top 10 universities\n4 out of ten universities')
plt.show()
```
